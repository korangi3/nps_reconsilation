```python
import pandas as pd
import re
```

```python
bank_path = r"D:\For Auto Recon\BANK STATEMENT REPORT FOR RECONCILE Phase1-20230807T053815Z-001\BANK STATEMENT REPORT FOR RECONCILE Phase1\ADBL 16-07-2020 to 15-07-2021.xlsx"
soa_path = r"D:\For Auto Recon\SOA REPORT FOR RECONCILE Phase1-20230807T053702Z-001\SOA REPORT FOR RECONCILE Phase1\Commission Report (07_16_2020 to 07_15_2021)ADBL.csv"
```

```python
bank_statement_df = pd.read_excel(bank_path )
soa_statement_df = pd.read_csv(soa_path,encoding='latin-1')
```

```python
bank_statement_df.head(15)
```

```python
bank_statement_df.drop(index=0,inplace=True)
```

```python
bank_statement_df.head()
```

```python
bank_statement_df['Narrative'].count()
```

```python
for index, row in bank_statement_df.iterrows():
    if "100000" in str(row["Narrative"]):
        id_matches = re.findall(r'10*[0-9]{6}', str(row["Narrative"]))
        if id_matches:
            id = id_matches[0].replace("100000", "")
            bank_statement_df.loc[index, 'Transaction Id'] = id
    soa_remarks = row['SOA Remark']
    if pd.notna(soa_remarks):  # Check if 'SOA Remarks' is not NaN
        bank_statement_df.at[index, 'Transaction Id'] = soa_remarks

bank_statement_df.head()
```

```python
bank_statement_df['Transaction Id'].count()
```

```python
test = bank_statement_df[bank_statement_df['SOA Remark'].notnull()]
```

```python
for index, row in test.iterrows():
    soa_remarks = row['SOA Remark']
    if pd.notna(soa_remarks):  # Check if 'SOA Remarks' is not NaN
        test.at[index, 'Transaction Id'] = soa_remarks
```
