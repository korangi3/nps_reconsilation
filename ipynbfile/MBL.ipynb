```python
bank_statement_df = bank_statement_df[
    (~bank_statement_df['Txn Reference'].str.contains('TRTR')) &
    (~bank_statement_df['Txn Reference'].str.contains('TRRR'))
]
```

```python
for index, row in bank_statement_df.iterrows():
    if '100000' in row['Txn Reference']:
    # Split the description into words
        description_words = row['Txn Reference'].split("/")
        value = description_words[0]
        id = re.sub(r'10*', '', value, count=1)
        bank_statement_df.loc[index, 'Transaction Id'] = id
    #     elif len(description_words) < 2:
    #         value = description_words[0]
    #         id = re.sub(r'10*', '', value, count=1)
    #         bank_statement_df.loc[index, 'Transaction Id'] = id

    # elif 'CIPS' in row['Description']:
    #     description_cips = row['Description'].split("/")
    #     value = description_cips[1]
    #     bank_statement_df.loc[index,'Transaction Id'] = value
```

```python
def standard_format(row):
        if row['Txn Amount'] == 'D' :
            mode = 'DR'
            amount = abs(row['Txn Amount'])
        elif row['Part Txn Type'] == 'C':
            mode = 'CR'
            amount = row['Txn Amount']
        else:
            mode = None
            amount = None
        return pd.Series([mode, amount], index=['Mode', 'Amount'])
bank_statement_df[['Mode', 'Amount']] = bank_statement_df.apply(standard_format, axis=1)  
```
