```python
for index, row in bank_statement_df.iterrows():
    # Check if 'Description' column has spaces
    # if not (' ' in row['Description']):
    #     continue
    if '100000' in row['Description']:
    # Split the description into words
        description_words = row['Description'].split("-")
        # print(description_words)

        # Check if there are at least 3 words (index 0, 1, 2)
        if len(description_words) >= 3:
            value = description_words[1]
            if '1000' in value:
                id = re.sub(r'10*', '', value, count=1)
                bank_statement_df.loc[index, 'Transaction Id'] = id
        elif len(description_words) < 2:
            value = description_words[0]
            id = re.sub(r'10*', '', value, count=1)
            bank_statement_df.loc[index, 'Transaction Id'] = id

    elif 'CIPS' in row['Description']:
        description_cips = row['Description'].split("/")
        value = description_cips[1]
        bank_statement_df.loc[index,'Transaction Id'] = value
```

```python
for index, row in bank_statement_df.iterrows():
        row_desc:str = row["Description"]
        id_text_list:list = row_desc.split()

        if "10000" in str(row_desc):
            id = re.sub(r'10*', '', id_text_list[-1], count=1)
            bank_statement_df.loc[index, 'Transaction Id'] = id
        elif 'CIPS' in row['Description']:
            description_cips = row['Description'].split("/")
            value = description_cips[1]
            bank_statement_df.loc[index,'Transaction Id'] = value

        elif 'IMEPAY' in str(row_desc) and not('TRFO' in str(row_desc) or 'TRFB' in str(row_desc)):
            id_text_list = row_desc.split('-')
            for id in id_text_list:
                if not re.match(r'^202', id):
                    continue
                bank_statement_df.loc[index, 'Transaction Id'] = id
        elif 'IMEPAY' in str(row_desc) and ('TRFO' in str(row_desc) or 'TRFB' in str(row_desc)):
            id_text_list = row_desc.split('-')
            # print('IMEPAY_ID==>',id_text_list)
            for id in id_text_list:
                # print('ID==>',id)
                if not re.match(r'^202', row_desc):
                    print('id==>',id)
                    continue
                
                bank_statement_df.loc[index, 'Transaction Id'] = id
        else:
            continue
```
