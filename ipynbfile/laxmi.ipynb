```python
# bank_statement_df = bank_statement_df.loc[bank_statement_df['TRAN_PARTICULAR'] != 'TRTR']
bank_statement_df = bank_statement_df[~bank_statement_df['TRAN_PARTICULAR'].str.contains('TRTR')]
bank_statement_df = bank_statement_df[~bank_statement_df['TRAN_PARTICULAR'].str.contains('TRRR')]
# bank_statement_df.drop(bank_statement_df[bank_statement_df['TRAN_PARTICULAR'] == 'TRTR/425160833773/25-04-2021'].index, inplace=True)
```

```python
for index, row in bank_statement_df.iterrows():
    if ('connectIPS' in row['TRAN_PARTICULAR']):
        tid = row['TRAN_PARTICULAR'].split('/')[-1]
        # print(tid)
        bank_statement_df.loc[index,'Transaction Id'] = tid
    if ( ',' in row['TRAN_PARTICULAR']):
        value = row['TRAN_PARTICULAR'].split(',')[-1]
        id = re.sub(r'10*', '', value, count=1)
        bank_statement_df.loc[index, 'Transaction Id'] = id

```

```python
for index1, bank_statement_soa in soa_statement_df.iterrows():
    transaction_id = bank_statement_soa['Transaction Id']

    if transaction_id in bank_statement_df['Transaction Id'].values:
        soa_statement_df.loc[index1,'Matched_tid'] = transaction_id
        continue
    soa_statement_df.loc[index1,'Unmatched_tid'] = transaction_id

print(soa_statement_df['Matched_tid'].count())
print(soa_statement_df['Unmatched_tid'].count())
```
