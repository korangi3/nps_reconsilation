```python
def process_bank_statement(bank_statement_df):
    def extract_transaction_id(row, col1, col2):
        for item in (row[col1], row[col2]):
            if "NPS-IF-" in str(item):
                id = re.findall(r'[0-9]{7}', item)
                return id[0]
            elif "10000" in str(item):
                id = re.findall(r'10*[0-9]{7}', item)
                return id[0].replace("10000", "")
            elif "FTMS-" in str(item):
                id = re.findall(r'[0-9]{6}', item)
                return id[0]
            # elif "eod" in item:
            #     return item
        return None

    bank_statement_df['Transaction id'] = bank_statement_df.apply(lambda row: extract_transaction_id(row, 'Desc1', 'Desc3'), axis=1)
    return bank_statement_df

# Example usage:
statement_df = process_bank_statement(df_statement)
statement_df.head()
```
