```python
import pandas as pd
import re
```

```python
bank_statement_df = pd.read_excel('D:/NPS/Manjushree.xlsx', header=0)
soa_statement_df = pd.read_excel('D:/NPS/Manjushree SOA.xlsx', header=0)
```

```python
for index, row in bank_statement_df.iterrows():
    # print(bank_statement_df.iloc[index]['transaction_id'])
        
        if "NPS-IF-" in str(row["Desc2"]):
            id = re.findall(r'[0-9]{7}',str(row["Desc2"]))
            bank_statement_df.loc[index, 'Transaction ID'] = id[0]
   
        elif "10000" in str(row["Desc2"]):
            id = re.findall(r'10*[0-9]{7}',str(row["Desc2"]))
            bank_statement_df.loc[index, 'Transaction ID'] = id[0].replace("10000","")
   
    
            
        
            
            
            
```
