```python
import pandas as pd
import re
```

```python
bank_statement_df = pd.read_excel(r"D:\NPS_phase1\output\BANK STMT PHASE 4\Bank_NMB.xlsx", skiprows=1)
# soa_statement_df = pd.read_excel('D:/NPS/NMB bank.xlsx', sheet_name='SOA')
```

```python
bank_statement_df['Merged_Description'] = bank_statement_df['Desc1'].astype(str) + bank_statement_df['Desc3']
```

```python
bank_statement_df.drop(columns=['Desc1','Desc3'],inplace=True)
# df = df.drop(columns=columns_to_drop)
```

```python
bank_statement_df.head()
```

```python
for index, row in bank_statement_df.iterrows():
    if 'NPS-IF-' in str(row['Merged_Description']):
        match = re.search(r'NPS-IF-(\d{7})', str(row["Merged_Description"]))
        if match:
            id = match.group(1)
            bank_statement_df.loc[index, 'Transaction Id'] = id

    elif "10000" in str(row["Merged_Description"]):
        id = re.findall(r'10000*[0-9]{7}',str(row["Particulars"]))
        bank_statement_df.loc[index, 'Transaction Id'] = id[0].replace("10000","")
    
    elif "FTMS-" in str(row["Merged_Description"]):
        match = re.search(r'FTMS-(\d+)',str(row["Merged_Description"]))
        if match:
            id= match.group(1)
            bank_statement_df.loc[index, 'Transaction Id'] = id
```

```python
bank_statement_df['Date'] = pd.to_datetime(bank_statement_df['Date'], format="%Y-%m-%dT%H:%M:%S").dt.date
```

```python
bank_statement_df.head()
```

```python
def standard_format_phase_4(row):
        """
        Determine transaction mode and amount from bank statement data.

        Parameters:
            row (Series): A row of bank statement data.

        Returns:
            Series: Mode (CR/DR) and amount.
        """

        if row['Txn Type'] == 'D' :
            mode = 'DR'
            amount = abs(row['Amount'])
        elif row['Txn Type'] == 'C':
            mode = 'CR'
            amount = abs(row['Amount'])
        else:
            mode = None
            amount = None
        return pd.Series([mode, amount], index=['Mode', 'Amount'])
```

```python
bank_statement_df[['Mode', 'Amount']] = bank_statement_df.apply(standard_format_phase_4, axis=1)
```

```python
excel_file_name = 'data.xlsx'

# Convert the DataFrame to an Excel file
bank_statement_df.to_excel(excel_file_name, index=False)
```
