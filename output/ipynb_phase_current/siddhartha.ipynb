```python
import pandas as pd
import re
```

```python
df_bank_stmt = pd.read_excel('D:\\NPS\\Siddhartha Bank STMT.xlsx')
```

```python
DR_amount = df_bank_stmt['DEBIT_AMOUNT'].sum()
CR_amount = df_bank_stmt['CREDIT_AMOUNT'].sum()
print(f'DR_amount:{DR_amount}')
print(f'CR_amount:{CR_amount}')
```

```python
for index, row in df_bank_stmt.iterrows():
    # print(bank_statement_df.iloc[index]['transaction_id'])
    for roww in (row['RESERVEDFIELD2'] , row['RESERVEDFIELD1']):
        
        if "NPSIF" in roww:
            id = re.findall(r'NPSIF(\d{7})',roww)
            df_bank_stmt.loc[index, 'Transaction id'] = id[0]
        
        elif "10000" in roww:
            id = re.findall(r'10*[0-9]{7}',roww)
            df_bank_stmt.loc[index, 'Transaction id'] = id[0].replace("10000","")
            
        elif "FTMS" in roww:
            id = re.findall(r'FTMS-(\d{6})',roww)
            df_bank_stmt.loc[index, 'Transaction id'] = id[0]
        
        elif re.findall(r"'[0-9]{7}",str(roww)):
            id = re.findall(r'[0-9]{7}',roww)
            df_bank_stmt.loc[index, 'Transaction id'] = id[0]
         

            
         
            
```
