```python
# import requests
from bs4 import BeautifulSoup
import pandas as pd

url = r"D:\NPS_phase1\nps-reconciliation\output\BANK STMT PHASE 4\Bank_NIBL.xls"

with open(url, 'r', encoding='utf-8') as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')

# print(f"worksheet_url===={soup}")

table = soup.find('table')

file.close()

bank_statement_df = pd.read_html(str(table), skiprows=2)[0]

bank_statement_df
```

```python
# Set the values in the first row as column names
bank_statement_df.columns = bank_statement_df.iloc[0]
# Drop the first row (which is now the column names)
bank_statement_df = bank_statement_df.iloc[1:]
```

```python
bank_statement_df
```

```python
for index, row in bank_statement_df.iterrows():
    if "NPS-IF-" in str(row["Description"]):
        match = re.search(r'NPS-IF-(\d{7})', str(row["Description"]))
        if match:
            id = match.group(1)
            bank_statement_df.loc[index, 'Transaction ID'] = id

    elif "10000" in str(row["Description"]):
        id = re.findall(r'10000*[0-9]{7}',str(row["Description"]))
        bank_statement_df.loc[index, 'Transaction ID'] = id[0].replace("10000","")
    
    elif "FTMS-" in str(row["Description"]):
        match = re.search(r'FTMS-(\d+)',str(row["Description"]))
        if match:
            id= match.group(1)
            bank_statement_df.loc[index, 'Transaction ID'] = id
```

```python
bank_statement_df['Transaction Amount'] = bank_statement_df['Transaction Amount'].astype(float)
```

```python
def standard_format_phase_4(row):
    """
    Determine transaction mode and amount from bank statement data.

    Parameters:
        row (Series): A row of bank statement data.

    Returns:
        Series: Mode (CR/DR) and amount.
    """
    if row['Cr/Dr'] == 'Dr' :
        mode = 'DR'
        amount = abs(row['Transaction Amount'])
    elif row['Cr/Dr'] == 'Cr':
        mode = 'CR'
        amount = row['Transaction Amount']
    else:
        mode = None
        amount = None
    return pd.Series([mode, amount], index=['Mode', 'Amount'])
```

```python
bank_statement_df[['Mode', 'Amount']] = bank_statement_df.apply(standard_format_phase_4, axis=1)
bank_statement_df.drop(columns=['Cr/Dr','Transaction Amount'],axis=1,inplace=True)
```
