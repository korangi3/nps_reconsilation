```python
bank_statement_df.drop(bank_statement_df.index[0],inplace=True)
bank_statement_df = bank_statement_df.iloc[:-3]
```

```python
bank_statement_df
```

```python
for index, row in bank_statement_df.iterrows():
    if "NPS-IF-" in str(row["Transaction Detail"]):
        id = re.findall(r'[0-9]{7}',str(row["Transaction Detail"]))
        bank_statement_df.loc[index, 'Transaction ID'] = id[0]

    elif "10000" in str(row["Transaction Detail"]):
        id = re.findall(r'10000*[0-9]{7}',str(row["Transaction Detail"]))
        bank_statement_df.loc[index, 'Transaction ID'] = id[0].replace("10000","")
    
    elif "FTMS-" in str(row["Transaction Detail"]):
        match = re.search(r'FTMS-(\d+)',str(row["Transaction Detail"]))
        if match:
            id= match.group(1)
            bank_statement_df.loc[index, 'Transaction ID'] = id
       
       
```
