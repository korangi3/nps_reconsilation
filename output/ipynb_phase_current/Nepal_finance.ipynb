```python
import pandas as pd
import re
```

```python
bank_statement_df = pd.read_excel(r"D:\NPS_phase1\nps-reconciliation\output\BANK STMT PHASE 4\Bank_Nepal_Finance.xlsx",skiprows=1)
```

```python
bank_statement_df
```

```python
bank_statement_df = bank_statement_df.loc[bank_statement_df['Desc1'] != '~Date summary']
bank_statement_df['Desc2'] = bank_statement_df['Desc2'].astype(str)
bank_statement_df['Desc2'] = bank_statement_df['Desc2'].str.rstrip('.0')
```

```python
bank_statement_df
```

```python
for index, row in bank_statement_df.iterrows():
    if '10000' in str(row['Desc2']):
        tran_id = str(row['Desc2']).split()
        value = tran_id[0]
        id = re.sub(r'10*', '', value, count=1)
        # print(id)
        bank_statement_df.loc[index, 'Transaction Id'] = id
```

```python
bank_statement_df
```

```python
bank_statement_df['Date'] = pd.to_datetime(bank_statement_df['TranDate'],format="%d/%m/%Y").dt.date
```

```python
bank_statement_df
```

```python
def standard_format(row):
        """
        Determine transaction mode and amount from bank statement data.

        Parameters:
            row (Series): A row of bank statement data.

        Returns:
            Series: Mode (CR/DR) and amount.
        """
        if row['Debit'] > 0:
            mode = 'DR'
            amount = row['Debit']
        elif row['Credit'] > 0:
            mode = 'CR'
            amount = row['Credit']
        else:
            mode = None
            amount = None
        return pd.Series([mode, amount], index=['Mode', 'Amount'])
```

```python
bank_statement_df[['Mode', 'Amount']] = bank_statement_df.apply(standard_format, axis=1)   
```
