```python
import pandas as pd
import re
```

```python

bank_statement_df = pd.read_excel(r"D:\NPS\phase3\phase3\ADBL.xls")
```

```python
for index, row in bank_statement_df.iterrows():
    if "NPS-IF-" in str(row["Narration"]):
        id = re.findall(r'[0-9]{7}',str(row["Narration"]))
        bank_statement_df.loc[index, 'Transaction Id'] = id[0]

    elif "10000" in str(row["Narration"]):
        id = re.findall(r'10000*[0-9]{7}',str(row["Narration"]))
        bank_statement_df.loc[index, 'Transaction Id'] = id[0].replace("10000","")
    
    elif "FTMS-" in str(row["Narration"]):
        match = re.search(r'FTMS-(\d+)',str(row["Narration"]))
        if match:
            id= match.group(1)
            bank_statement_df.loc[index, 'Transaction Id'] = id
    elif "FT-" in str(row["Narration"]):
        match = re.search(r'FT-(\d+)',str(row["Narration"]))
        if match:
            id= match.group(1)
            bank_statement_df.loc[index, 'Transaction Id'] = id
       
       
```

```python
def standard_format(row):
        """
        Standardize transaction amount and mode.

        Classify transactions as debit (DR) or credit (CR).
        """
        if abs(row['DEBIT']) > 0:
            mode = 'DR'
            amount = row['DEBIT']
        elif row['CREDIT'] > 0:
            mode = 'CR'
            amount = row['CREDIT']
        else:
            mode = None
            amount = None
        return pd.Series([mode, amount], index=['Mode', 'Amount'])
```

```python
bank_statement_df[['Mode', 'Amount']] = bank_statement_df.apply(standard_format, axis=1)   
```
